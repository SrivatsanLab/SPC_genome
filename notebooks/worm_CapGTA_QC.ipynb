```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pal = {
    'unsync_10_ns':'#03FFFF',
    'L4_19_nova':'#E980FC'
}
```

```python
unsync_10_ns_qc = pd.read_csv('../results/worm_CapGTA_pilot/PolE_worm_pilot_qc_summary.csv', index_col='sample')
unsync_10_ns_qc['experiment'] = 'unsync_10_ns'
```

```python
L4_19_nova_qc = pd.read_csv('../results/worm_CapGTA_UDI_5/sc_qc/all_cells_dna_qc_metrics.csv', index_col='sample')
L4_19_nova_qc['experiment'] = 'L4_19_nova'
```

```python
data = pd.concat([unsync_10_ns_qc, L4_19_nova_qc])
```

```python
data
```

```python
# sns.set_theme(style="ticks", palette="pastel")
fig,ax = plt.subplots(1,1,figsize=(10,6))
sns.boxplot(data=data, x="experiment", y="mean_coverage",
            ax=ax,
            palette=pal,
            whis=(0,100),
            width=.6
           )

sns.stripplot(data=data, x="experiment", y="mean_coverage",
              ax=ax,
              size=4, 
              color=".25"
             )

sns.despine()
plt.xlabel('')
plt.yscale('log')
plt.ylabel('Mean Coverage')
```
